```python
import pandas as pd
```

```python
data = pd.read_csv('../Orginal_data/billboard.csv', encoding='ISO-8859-1')
data
```

```python
mod_data = data.sort_values(['track'], ascending=True)  #sorting data by track name
mod_data = mod_data.reset_index(drop=True) #reseting index
mod_data
```

```python
mod_data["date.entered"] = pd.to_datetime(mod_data["date.entered"])  #change column type to date
mod_data["date.peaked"] = pd.to_datetime(mod_data["date.peaked"])
mod_data.describe()
```

```python
artist_tracks = mod_data[['artist.inverted', 'track', 'time', 'genre']]
artist_tracks = artist_tracks.rename(columns={'artist.inverted' : 'Artist', 'track' : 'Track', 'time' : 'Time', 'genre' : 'Genre'})
artist_tracks.head(7)
```

```python
artist_tracks.to_csv('../Analysis_data/artists_tracks.csv')
```

```python
ranking = mod_data.dropna(axis=1)
ranking = ranking.drop(columns=['year'])
ranking['date.peaked'] = ranking['date.peaked'].sub(ranking['date.entered'], axis=0)
ranking = ranking.drop(['date.entered'], axis=1)
ranking = ranking.rename(columns={'artist.inverted' : 'Artist', 'track' : 'Track', 'genre': 'Genre', 'x1st.week' : 'Ranking position', 'date.peaked' : 'Ranking time'})
ranking = ranking.sort_values(['Ranking position'], ascending=True)
ranking = ranking.reset_index(drop=True)

ranking
```
